```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/train.csv')
```

```python
df.head()
```

Scatterplot: Numerical Numerical

```python
sns.scatterplot(data=df,x='ApplicantIncome',y='LoanAmount',hue=df['Gender'],style=df['Married'],size=df['Property_Area'])
```

Numerical - Categorical(Barplot)

```python
sns.barplot(data=df,x='Gender',y='LoanAmount',hue='Married')
```

Boxplot

```python
sns.boxplot(data=df,x='Credit_History',y='LoanAmount',hue='Self_Employed')
```

Distplot

```python
sns.distplot(df[df['Loan_Status']=='Y']['ApplicantIncome'],hist=False)
sns.distplot(df[df['Loan_Status']=='N']['ApplicantIncome'],hist=False)
```

Categorical Categorical

```python
sns.heatmap(pd.crosstab(df['Education'],df['Loan_Status']))
```

```python
sns.heatmap(pd.crosstab(df['Credit_History'],df['Loan_Status']))
```

```python
sns.heatmap(pd.crosstab(df['Property_Area'],df['Loan_Status']))
```

```python
df['LoanStatus_Numeric'] = df['Loan_Status'].apply(lambda x: 1 if x == 'Y' else 0)

# Calculate the mean of the numerical 'LoanStatus' and plot
(df.groupby('Gender')['LoanStatus_Numeric'].mean()*100).plot(kind='bar')
```

Clustermap

```python
sns.clustermap(pd.crosstab(df['Dependents'],df['Loan_Status']))
```

Pairplot

```python
sns.pairplot(df,hue='Education')
```

Line plot: for time series analysis
