```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/covid_related_disease_data.csv')
```

```python
df.head()
```

Categorical Data

1.Countplot

```python
sns.countplot(df['Gender'])
```

```python
sns.countplot(df['COVID_Strain'])
```

```python
df['Smoking_Status'].value_counts().plot(kind='bar')
```

2. Piechart

```python
df['COVID_Strain'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
df['Occupation'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
df['Vaccine_Type'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
df['Reinfection'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```
